```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

%matplotlib inline
```

```python
df=pd.read_csv("processed_crime_data.csv")
```

```python
df1=df.groupby(by=['CAT']).size().sort_values(ascending=False)
df1.plot.bar(x='category',ylabel='count',grid=False,figsize=(3,3))
```

```python
df2=df.groupby(by=['YEAR']).size()
df2.plot.line(x='Year',ylabel='count',grid=True,figsize=(5,5))
```
